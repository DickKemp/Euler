## Problem 1 - Multiples of 3 and 5

If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

```python
s =0
for i in range (1,1000):
    if i % 3 == 0 or i % 5 == 0:
        s += i

print s
```

## Problem 3 - Largest prime factor

The prime factors of 13195 are 5, 7, 13 and 29.

What is the largest prime factor of the number 600851475143 ?

To find the prime factors of a number, we first need a way to find the prime numbers

```python
def clearMultipliers(arr, idx, multiplier, end):
    if idx < 0:
        return arr
    idx += multiplier
    while idx <= end:
        arr[idx] = 0
        idx += multiplier
    return arr
 
def findNextNonZero(arr, idx, end):
    idx += 1
    while idx <= end:
        if arr[idx] != 0:
            return idx
        idx += 1
    return -1
 
def findPrimesLEThan(N):
    arr = list(range(N+1))
    primes = []
    p = 1
    while p > 0:
        p = findNextNonZero(arr, p, N)
        if (p > 0):     
            clearMultipliers(arr, p, p, N)
            primes.append(p)
    return primes
```

### simple test of the prime numbers less than 100

```python
print(findPrimesLEThan(100))
```

### Finding the prime factors of a number is done as follows
Find all of the prime numbers less than the square root of the number
Iterate through those prime numbers and check if it is a factor of the number
If it is, then store it away as a prime factor and apply that factor to the number to reduce for the next iteration

This will work so long as we have enough memory to compute the primes of the square root of the number in question

```python
import math
def findPrimeFactors(num):
    primeFactors = []
    q = num
    for p in findPrimesLEThan(int(math.sqrt(num))):
        if q % p == 0:
            primeFactors.append(p)
            q = q / p
    return primeFactors

def findMaxPrimeFactor(num):
    return max(findPrimeFactors(num))

print findMaxPrimeFactor(600851475143)
```

## Problem 5 - Smallest multiple

2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

```python
def multiply(nums):
    p = 1
    for n in nums:
        p *= n
    return p
```

```python

# return true if q is cleanly divisible by all numbers in the list L
def verifyCanDivideList(q, L):
    for i in L:
        if q % i != 0:
            return False
    return True

# return true if q is cleanly divisible by all numbers from 1 to n
def verifyCanDivide(q, n):
    return verifyCanDivideList(q,range(1,n+1))
```

The following is a brute force algorithm which works for n = 10, but not for larger values such as 20

```python
def bruteForceSmallestMultiple(n):
    low = multiply(findPrimesLEThan(n))
    high = multiply(range(1,n+1))
    for i in xrange (low,high):
        if verifyCanDivide(i, n):
            return i
    return -1
bruteForceSmallestMultiple(10)
```

```python
def findNonPrimesLTE(n):
    p = findPrimesLEThan(n)
    np = [x for x in range(4, n+1) if x not in p]
    return np
print(findPrimesLEThan(20))
print(findNonPrimesLTE(20))

import itertools
def findsubsets(S,m):
    return set(itertools.combinations(S, m))
S = [4, 6, 8, 9]
for t in [list(t) for t in findsubsets(S,3)]:
    print t
```

```python
import sys
def smallestMultiple(n):
    low = multiply(findPrimesLEThan(n))
    otherMultiples = range(2,n)
    lowest = sys.maxint
    count = 0
    for i in xrange (1,len(otherMultiples)):
        for r in [list(r) for r in findsubsets(otherMultiples, i)]:
            num = low * multiply(r)
            if (num > lowest):
                continue
            count += 1
            if verifyCanDivideList(num, otherMultiples):
                if num < lowest:
                    lowest = num
    print("# tested: {0}".format(count))
    return lowest

smallestMultiple(20)
```

## Problem 6 - Sum square difference

The sum of the squares of the first ten natural numbers is,

$$1^{2} + 2^{2} + ... + 10^{2} = 385$$
The square of the sum of the first ten natural numbers is,

$$(1 + 2 + ... + 10)^{2} = 55^{2} = 3025$$

Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is 3025 âˆ’ 385 = 2640.

Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.

```python
import math
sumOfSquares = 0
theSum = 0
for i in xrange(1, 101):
    sumOfSquares += math.pow(i,2)
    theSum += i
print (int(math.pow(theSum, 2) - sumOfSquares))
```

## Problem 51 - Prime digit replacements

By replacing the 1st digit of the 2-digit number *3, it turns out that six of the nine possible values: 13, 23, 43, 53, 73, and 83, are all prime.

By replacing the 3rd and 4th digits of 56**3 with the same digit, this 5-digit number is the first example having seven primes among the ten generated numbers, yielding the family: 56003, 56113, 56333, 56443, 56663, 56773, and 56993. Consequently 56003, being the first member of this family, is the smallest prime with this property.

Find the smallest prime which, by replacing part of the number (not necessarily adjacent digits) with the same digit, is part of an eight prime value family.

```python
primes = findPrimesLEThan(10000000)
lenCounts = {}
biggest = primes[len(primes)-1]
for j in range(1,len(str(biggest))+1):
    lenCounts[j] = 0
for i in primes:
    pLen = len(str(i))
    lenCounts[pLen] += 1

for k in lenCounts:
    print("{0} is {1}".format(k, lenCounts[k]))
```
